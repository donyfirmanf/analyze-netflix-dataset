```python
import matplotlib.pyplot as plt

# Bar chart of the number of movies per rating
df['rating'].value_counts().sort_values().plot(kind='bar', color="darkcyan", edgecolor='black', alpha=0.7)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i in range(len(df['rating'].value_counts().sort_values())):
    count = df['rating'].value_counts().sort_values().values[i]
    plt.annotate(count, xy=(i, count), xytext=(0, 3), textcoords="offset points", ha='center')
    
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Rating')
plt.ylim(0,3000)
plt.xticks(rotation=45)
plt.legend(['Movies'])
plt.show()


```

```python
# Bar chart of the number of movies per type
df['type'].value_counts().sort_values().plot(kind='bar', color='green', edgecolor='black', alpha=0.7)

for i in range(len(df['type'].value_counts().sort_values())):
    count = df['type'].value_counts().sort_values().values[i]
    plt.annotate(count, xy=(i, count), xytext=(0, 3), textcoords="offset points", ha='center')

plt.xlabel('Type')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Type')
plt.xticks(rotation=45)
plt.legend(['Movies'])
plt.show()
```
